```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python

theta = (2 * math.pi) / 3
corners = [[math.cos(i * theta), math.sin(i * theta)] for i in range(3)]

offset = math.pi / 3
seg_len = -math.cos(1 * theta)
midpoints = [[seg_len * math.cos(i * theta + offset), seg_len * math.sin(i * theta + offset)] for i in range(3)]
normals = [corners[(i + 2) % 3] for i in range(3)]

for p in corners:
    print(p)

for m in midpoints:
    print(m)

for n in normals:
    print(n)

plt.scatter([p[0] for p in corners], [p[1] for p in corners])
plt.scatter([m[0] for m in midpoints], [m[1] for m in midpoints])
plt.axis('scaled')
plt.show()

def is_inside_triangle(p):
    for i in range(3):
        d = p - midpoints[i]
        if np.dot(d, normals[i]) <= 0:
            return False
    return True

def generate_data(num_items):
    points = np.array([np.random.uniform(-1, 1, num_items),np.random.uniform(-1, 1, num_items)]).transpose()
    inside = [is_inside_triangle(p) for p in points]
        
    return points, inside

ppp, inside = generate_data(10000)
plt.scatter([p[0] for p in ppp], [p[1] for p in ppp], c=inside)
plt.axis('scaled')
plt.show()
```
