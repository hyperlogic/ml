```python
import numpy as np
import matplotlib.pyplot as plt

def plane(x, y, mx=0, my=0, b=0, noise=0):
    return mx * x + my * y + b + (noise * np.random.rand(1)[0])

# nieve linear regression, just apply the formula
def my_linear_regression(X, y):
    return np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)
```

```python
# create the X matrix
NUM_SAMPLES = 10
xx = np.linspace(0, 1, NUM_SAMPLES)
yy = np.linspace(0, 1, NUM_SAMPLES)
ones = np.ones(shape=(1, NUM_SAMPLES * NUM_SAMPLES))
XX, YY = np.meshgrid(xx, yy)
X = np.vstack([XX.ravel(), YY.ravel(), ones]).T
y = np.zeros(shape=(NUM_SAMPLES * NUM_SAMPLES, 1))
for i in range(NUM_SAMPLES * NUM_SAMPLES):
    y[i] = plane(X[i,0], X[i,1], mx=0, my=0.25, b=0, noise=0.2)

w = my_linear_regression(X, y)


fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.scatter(X[:,0], X[:,1], y)
ax.scatter(X[:,0], X[:,1], X.dot(w), color='red')
plt.axis('scaled')
plt.show()

w
```
