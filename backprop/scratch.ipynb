```python
import torch
import torchviz
```

```python
x = torch.Tensor([1])
x.requires_grad = True

# y = (3x^3 + 1)(x^2+6x)
three = torch.Tensor([3])
six = torch.Tensor([6])

y = (three * torch.pow(x, 3) + torch.Tensor([1])) * (torch.pow(x, 2) + six * x)

torchviz.make_dot(y, params=dict(y=y,x=x,three=three), show_attrs=True, show_saved=True)
```
